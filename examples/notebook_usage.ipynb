# Using the Sentiment Analysis Model in a Notebook

This notebook demonstrates how to use the trained model directly in Python (without the API).

```python
# Add parent directory to path
import sys
sys.path.append('..')

from ml_pipeline.predict import load_model_and_vectorizer, predict_sentiment, predict_batch
```

```python
# Load model (only once)
model, vectorizer = load_model_and_vectorizer()
print("Model loaded successfully!")
```

```python
# Single prediction
review = "This movie was absolutely fantastic! I loved every minute."
result = predict_sentiment(review, model, vectorizer)
print(f"Text: {review}")
print(f"Sentiment: {result['sentiment']}")
print(f"Confidence: {result['confidence']:.4f}")
```

```python
# Batch prediction
reviews = [
    "Best movie ever!",
    "Worst film I've seen.",
    "It was okay, nothing special."
]

results = predict_batch(reviews, model, vectorizer)
for review, result in zip(reviews, results):
    print(f"\n'{review}'")
    print(f"  → {result['sentiment'].upper()} ({result['confidence']:.4f})")
```

## Analyzing Multiple Reviews

```python
import pandas as pd

# Sample reviews
data = {
    'review': [
        "Absolutely brilliant! A masterpiece.",
        "Terrible acting, boring story.",
        "The special effects were amazing but the plot was weak.",
        "I couldn't stop watching. Fantastic!",
        "What a waste of two hours."
    ]
}

df = pd.DataFrame(data)

# Predict
predictions = predict_batch(df['review'].tolist(), model, vectorizer)
df['sentiment'] = [p['sentiment'] for p in predictions]
df['confidence'] = [p['confidence'] for p in predictions]

df
```
